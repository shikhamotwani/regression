"Simple linear regression is a basic statistical method used to understand the relationship between two continuous variables. It assumes that there is a linear relationship between the independent variable (also known as the predictor or input variable) and the dependent variable (also known as the response or output variable). In simple linear regression, we aim to find the best-fitting line through the data points that minimizes the overall distance between the observed data points and the line. This line is represented by the equation y = mx + b, where y is the dependent variable, x is the independent variable, m is the slope of the line, and b is the y-intercept. The slope (m) indicates the change in the dependent variable for a one-unit change in the independent variable, while the y-intercept (b) represents the value of the dependent variable when the independent variable is zero. Simple linear regression is often used for prediction and understanding the relationship between variables in various fields, including economics, finance, and social sciences."







# learn linear regression with example

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("students.csv")
```

```python
df.head()
```

```python
plt.scatter(df['CGPA'], df['Package (per annum)'])
plt.xlabel('CGPA')
plt.ylabel('Package (per annum)')
plt.title('CGPA vs Package (per annum)')
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
x_test_instance = x_test.iloc[1].values.reshape(1, 1)
predicted_package = lr.predict(x_test_instance)
predicted_package_rounded = round(predicted_package[0])
print("Predicted package (rounded):", predicted_package_rounded)
```

```python
plt.scatter(df['CGPA'], df['Package (per annum)'], label='Actual Data')
predicted_package = lr.predict(df['CGPA'].values.reshape(-1, 1))
plt.scatter(df['CGPA'], predicted_package, color='red', label='Predicted Data')
plt.xlabel('CGPA')
plt.ylabel('Package (per annum)')
plt.title('CGPA vs Package (per annum)')
plt.legend()
plt.show()
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```
